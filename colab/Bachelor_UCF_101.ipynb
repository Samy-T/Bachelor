# Training



```
!python /mnt/Bachelor/main.py ucf101 RGB \
     --arch mobilenetv2 --num_segments 8 \
     --gd 20 --lr 0.01 --lr_steps 20 40 --epochs 25 \
     --batch-size 32 -j 4 --dropout 0.5 --consensus_type=avg --eval-freq=1 \
     --shift --shift_div=8 --shift_place=blockres --npb \
     --root_log '/content/drive/MyDrive/Turculescu/Train-ucf101/MobileNet_V2/Train/log' \
     --root_model '/content/drive/MyDrive/Turculescu/Train-ucf101/MobileNet_V2/Train/checkpoint'
```
